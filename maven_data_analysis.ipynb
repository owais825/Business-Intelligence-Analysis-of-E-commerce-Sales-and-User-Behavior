Traffic & Session-Level Exploration

```python
# Total Sessions & Time Range

website_sessions[['created_at']].agg(['min', 'max'])
```

```python
# Sessions Over Time (Daily Trend)
sessions_daily = (
    website_sessions
    .groupby(website_sessions['created_at'].dt.date)
    .agg(sessions=('website_session_id', 'nunique'))
    .reset_index()
)

sessions_daily.head()
```

```python
# Traffic Source Breakdown
traffic_source = (
    website_sessions
    .groupby(['utm_source', 'utm_campaign'])
    .agg(sessions=('website_session_id', 'nunique'))
    .reset_index()
    .sort_values('sessions', ascending=False)
)

traffic_source.head(10)
```

```python
# Device Type Distribution
device_distribution = (
    website_sessions
    .groupby('device_type')
    .agg(sessions=('website_session_id', 'nunique'))
    .reset_index()
)

device_distribution
```

Session-Order Conversion Analysis

```python
# Overall Conversion Rate

sessions = website_sessions['website_session_id'].nunique()
orders_count = orders['order_id'].nunique()

conversion_rate = orders_count / sessions

print('sessions =',sessions)
print('orders_count =', orders_count)
print('conversion_rate =', conversion_rate)
```

```python
# Conversion Rate by Traffic Source
conv_by_source = (
    website_sessions
    .merge(orders, on='website_session_id', how='left')
    .groupby(['utm_source', 'utm_campaign'])
    .agg(
        sessions=('website_session_id', 'nunique'),
        orders=('order_id', 'nunique')
    )
    .reset_index()
)

conv_by_source['conversion_rate'] = (
    conv_by_source['orders'] / conv_by_source['sessions']
)

conv_by_source.sort_values('sessions', ascending=False).head(10)
```

```python
# Conversion Rate by Device
conv_by_device = (
    website_sessions
    .merge(orders, on='website_session_id', how='left')
    .groupby('device_type')
    .agg(
        sessions=('website_session_id', 'nunique'),
        orders=('order_id', 'nunique')
    )
    .reset_index()
)

conv_by_device['conversion_rate'] = (
    conv_by_device['orders'] / conv_by_device['sessions']
)

conv_by_device
```

Funnel Analysis (Page-Level Drop-Offs)

```python
# Identify Key Funnel Pages (which pages exist and how popular they are)

page_counts = (
    website_pageviews
    .groupby('pageview_url')
    .agg(pageviews=('website_pageview_id', 'count'))
    .reset_index()
    .sort_values('pageviews', ascending=False)
)

page_counts
```

```python
# First Page per Session (Landing Page)

first_page_per_session = (
    website_pageviews
    .sort_values('created_at')
    .groupby('website_session_id')
    .first()
    .reset_index()
)

first_page_per_session[['website_session_id', 'pageview_url']].head()
```

```python
funnel = (
    website_pageviews
    .assign(
        landing=lambda x: (
            x['pageview_url'].eq('/home') |
            x['pageview_url'].str.startswith('/lander')
        ),
        product=lambda x: x['pageview_url'].str.startswith('/the-'),
        cart=lambda x: x['pageview_url'].eq('/cart'),
        shipping=lambda x: x['pageview_url'].eq('/shipping'),
        billing=lambda x: (
            x['pageview_url'].eq('/billing') |
            x['pageview_url'].eq('/billing-2')
        ),
        thank_you=lambda x: x['pageview_url'].eq('/thank-you-for-your-order')
    )
    .groupby('website_session_id')
    .agg(
        landing=('landing', 'max'),
        product=('product', 'max'),
        cart=('cart', 'max'),
        shipping=('shipping', 'max'),
        billing=('billing', 'max'),
        thank_you=('thank_you', 'max')
    )
    .reset_index()
)
```

```python
# Funnel Volume at Each Stage
funnel_counts = funnel[
    ['landing', 'product', 'cart', 'shipping', 'billing', 'thank_you']
].sum()

funnel_counts
```

```python
# Funnel Conversion Rates
funnel_rates = funnel_counts / funnel_counts.shift(1)
funnel_rates
```

Revenue, AOV & Profit Analysis

```python
# Overall Revenue & Profit

total_revenue = orders['price_usd'].sum()
total_cogs = orders['cogs_usd'].sum()
total_profit = total_revenue - total_cogs
profit_margin = total_profit / total_revenue

print("total revenue =",total_revenue)
print("total profit =",total_profit)
print("profit margin =",profit_margin)
```

```python
# Average Order Value (AOV)
aov = orders['price_usd'].mean()
aov
```

```python
# Revenue Over Time (Monthly Trend)
orders['order_month'] = orders['created_at'].dt.to_period('M')

monthly_revenue = (
    orders
    .groupby('order_month')
    .agg(
        revenue=('price_usd', 'sum'),
        orders=('order_id', 'nunique'),
        avg_order_value=('price_usd', 'mean')
    )
    .reset_index()
)

monthly_revenue.head()
```

```python
# Revenue per Session (Quality Metric)
revenue_per_session = total_revenue / website_sessions['website_session_id'].nunique()
revenue_per_session
```

Product Performance & Refund Impact Analysis

```python
order_items_enriched = (
    order_items
    .merge(products, on='product_id', how='left')
)
```

```python
order_items_enriched['profit'] = (
    order_items_enriched['price_usd'] - order_items_enriched['cogs_usd']
)
```

```python
product_performance = (
    order_items_enriched
    .groupby('product_name')
    .agg(
        units_sold=('order_item_id', 'count'),
        revenue=('price_usd', 'sum'),
        profit=('profit', 'sum')
    )
    .reset_index()
)

product_performance
```

```python
# Primary vs Cross-Sell Items
primary_vs_cross = (
    order_items
    .merge(products, on='product_id', how='left')
    .groupby(['product_name', 'is_primary_item'])
    .agg(units=('order_item_id', 'count'))
    .reset_index()
)

primary_vs_cross
```

```python
# Refund Impact by Product
refund_impact = (
    order_items_refund
    .merge(order_items, on='order_item_id', how='left')
    .merge(products, on='product_id', how='left')
    .groupby('product_name')
    .agg(
        refund_count=('order_item_refund_id', 'count'),
        refund_amount=('refund_amount_usd', 'sum')
    )
    .reset_index()
    .sort_values('refund_amount', ascending=False)
)

refund_impact
```

```python
# Net Revenue After Refunds
net_revenue = (
    product_performance
    .merge(refund_impact, on='product_name', how='left')
)

net_revenue['refund_amount'] = net_revenue['refund_amount'].fillna(0)
net_revenue['net_revenue'] = net_revenue['revenue'] - net_revenue['refund_amount']

net_revenue.sort_values('net_revenue', ascending=False)
```
